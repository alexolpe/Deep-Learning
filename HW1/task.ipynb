In my case, it is the Iterator class. Every time the \_\_next\_\_ method is called, it returns the next element of the array until reaching the end. This is done by increasing the value of self.index every time and accessing the position self.index in the array.

```python
class Sequence (object):
    def __init__ (self, array):
        self.array = array
        
    def __iter__(self):
        return Iterator(self)
    
    def __len__(self):
        return len(self.array)
    
class Iterator:
    def __init__(self, seq):
        self.items = seq.array
        self.index = -1
    def __iter__(self):
        return self
    def __next__(self):
        self.index += 1
        if self.index < len(self.items):
            return self.items[self.index]
        else:
            raise StopIteration

class Arithmetic (Sequence):
    def __init__(self, start, step):
        self.start = start
        self.step = step
        
    def __call__(self, length):
        self.array = list(range(self.start, self.start+length*self.step, self.step))
        #print(self.array)
```

5. I copy the structure of the Arithmetic class to create the Geometric class. I just change the way of creating the array that is stored in self.array once the start value, ratio and length are specified in order to create a geometric sequence instead of an arithmetic sequence. 

```python
class Geometric (Sequence):
    def __init__(self, start, ratio):
        self.start = start
        self.ratio = ratio
        
    def __call__(self, length):
        self.array = [self.start * (self.ratio ** i) for i in range(length)]
        #print(self.array)
```

6. In this case I define the method \_\_eq\_\_ in the superclass Sequence. This method will be called when using the operator ==. The arguments of the method \_\_eq\_\_ are "self" and "other". "self" refers to the object in the left side of the operator == and "other" refers to the object in the right side of the operator ==. Using the method \_\_len\_\_ previously defined, I first compare that the lengths are the same. If not, I throw a ValueError exception. Otherwise, equal_count stores the number of elements that are equal element_wise. I use "for a, b in zip(self, other)" to go through the elements of both arrays element-wise calling the \_\_iter\_\_ method previously defined. "if a == b" I generate "1" for the counter. Finally, all the 1s are summed.

```python
class Sequence (object):
    def __init__ (self, array):
        self.array = array
        
    def __iter__(self):
        return Iterator(self)
    
    def __len__(self):
        return len(self.array)
    
    def __eq__(self, other):
        if len(self) != len(other):
            raise ValueError("Two arrays are not equal in length !")
        equal_count = sum(1 for a, b in zip(self, other) if a == b) 
        return equal_count
    
class Iterator:
    def __init__(self, seq):
        self.items = seq.array
        self.index = -1
    def __iter__(self):
        return self
    def __next__(self):
        self.index += 1
        if self.index < len(self.items):
            return self.items[self.index]
        else:
            raise StopIteration

class Arithmetic (Sequence):
    def __init__(self, start, step):
        self.start = start
        self.step = step
        
    def __call__(self, length):
        self.array = list(range(self.start, self.start+length*self.step, self.step))
        #print(self.array)
        
class Geometric (Sequence):
    def __init__(self, start, ratio):
        self.start = start
        self.ratio = ratio
        
    def __call__(self, length):
        self.array = [self.start * (self.ratio ** i) for i in range(length)]
        #print(self.array)
```
